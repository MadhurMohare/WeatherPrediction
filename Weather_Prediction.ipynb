```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('seattle-weather.csv')
df.head()
```

```python
df.isnull().sum()
```

```python
def labelEncoding(c):
    from sklearn.preprocessing import LabelEncoder
    le = LabelEncoder()
    df[c] = le.fit_transform(df[c])
    df[c].unique
labelEncoding('weather')
df
```

```python
cols = ['precipitation', 'temp_max', 'temp_min', 'wind']
```

```python
def Normalize(df,cols):
  for x in cols:
    #df[x] = (df[x] - df[x].min()) / (df[x].max() - df[x].min())
    df[x] = df[x]/df[x].max()
Normalize(df,cols)
df
```

```python
df = df.drop(['date'], axis=1)
df
```

```python
x = df.drop(['weather'], axis=1)
y = df['weather']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)
```

```python
from xgboost import XGBClassifier
model = XGBClassifier()
model.fit(x_train, y_train)
```

```python
pred = model.predict(x_test)
print(pred)
```

```python
from sklearn.metrics import classification_report,accuracy_score
print(classification_report(y_test, pred))
print(accuracy_score(y_test, pred))
```

```python
grid = {'learning_rate': [0.1,1,0.01,0.001], 'gamma':[0,1,10,100] }
```

```python
from sklearn.model_selection import GridSearchCV
grid_search = GridSearchCV(model, grid, cv=10, verbose=2)
grid_result = grid_search.fit(x_train, y_train)
```

```python
new_pred = grid_result.predict(x_test)
print(new_pred)
```

```python
print(classification_report(y_test, new_pred))
print(accuracy_score(y_test, new_pred))
```
